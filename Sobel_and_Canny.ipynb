```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
img = cv2.imread('ii.jpg')
```

```python
cv2_imshow(img)
```

```python
blurred = cv2.GaussianBlur(img, (5, 5), 0)
```

```python
sobelx = cv2.Sobel(blurred, cv2.CV_64F, 1, 0, ksize=3)  # X direction
sobely = cv2.Sobel(blurred, cv2.CV_64F, 0, 1, ksize=3)  # Y direction
```

```python
sobel_magnitude = cv2.magnitude(sobelx, sobely)
sobel_magnitude = cv2.convertScaleAbs(sobel_magnitude)
```

```python
titles = ['Original', 'Sobel X', 'Sobel Y', 'Sobel Magnitude']
images = [img, sobelx, sobely, sobel_magnitude]

plt.figure(figsize=(12, 6))
for i in range(4):
    plt.subplot(2, 2, i+1)
    plt.imshow(images[i], cmap='gray')
    plt.title(titles[i])
    plt.axis('off')
plt.show()
```
